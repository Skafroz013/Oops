```python
#Problem 1: 
#Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. 
#Implement methods to deposit and withdraw money from the account.


class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance
    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            return f"Deposited ${amount}. New balance: ${self.balance}"
        else:
            return "Invalid deposit amount. Please enter a positive value."
    def withdraw(self, amount):
        if amount > 0:
            if amount <= self.balance:
                self.balance -= amount
                return f"Withdrew ${amount}. New balance: ${self.balance}"
            else:
                return "Insufficient funds for withdrawal."
        else:
            return "Invalid withdrawal amount. Please enter a positive value."
    my_account = BankAccount("123456789", "Afroz", 1000.0)
    print(my_account.deposit(500))
    print(my_account.withdraw(200))
    print(my_account.withdraw(1500))
```

```python
#Problem 2:
#Employee Management Create a class representing an employee with attributes like employee ID, name, and salary.
#Implement methods to calculate the yearly bonus and display employee details.

class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary
    def calculate_yearly_bonus(self, bonus_percentage):
        if bonus_percentage >= 0:
            yearly_bonus = (bonus_percentage / 100) * self.salary
            return f"Yearly bonus for {self.name}: ${yearly_bonus:.2f}"
        else:
            return "Invalid bonus percentage. Please enter a non-negative value."
    employee1 = Employee("E12345", "Likhitha", 60000.0)
    print(employee1.calculate_yearly_bonus(10))  # 10% bonus
    print(employee1.calculate_yearly_bonus(-5))  # Invalid negative bonus
```

```python
#Problem 3: 
#Vehicle Rental Create a class representing a vehicle rental system.
#Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}  
    def add_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            return f"Added {vehicle_type} (ID: {vehicle_id}) to available vehicles."
        else:
            return f"Vehicle with ID {vehicle_id} already exists in the system."
    def rent_vehicle(self, vehicle_id):
        if vehicle_id in self.available_vehicles:
            rented_vehicle_type = self.available_vehicles.pop(vehicle_id)
            return f"Rented {rented_vehicle_type} (ID: {vehicle_id})."
        else:
            return f"Vehicle with ID {vehicle_id} is not available for rent."
    def return_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            return f"Returned {vehicle_type} (ID: {vehicle_id})."
        else:
            return f"Vehicle with ID {vehicle_id} is already available."
    def display_available_vehicles(self):
        if self.available_vehicles:
            available_vehicles_list = "\n".join(
                [f"ID: {id}, Type: {vehicle_type}" for id, vehicle_type in self.available_vehicles.items()]
            )
            return f"Available Vehicles:\n{available_vehicles_list}"
        else:
            return "No vehicles are currently available."
    rental_system = VehicleRentalSystem()
    print(rental_system.add_vehicle("V001", "Car"))
    print(rental_system.add_vehicle("V002", "Bike"))
    print(rental_system.add_vehicle("V003", "Truck"))

    # Displaying available vehicles
    print(rental_system.display_available_vehicles())

    # Renting a vehicle
    print(rental_system.rent_vehicle("V001"))

    # Displaying available vehicles after renting
    print(rental_system.display_available_vehicles())

    # Returning a vehicle
    print(rental_system.return_vehicle("V001", "Car"))

    # Displaying available vehicles after returning
    print(rental_system.display_available_vehicles())
```

```python
#Problem 4: 
#Library Catalog Create classes representing a library and a book. 
#Implement methods to add books to the library, borrow books, and display available books.

class Book:
    def __init__(self, title, author, book_id):
        self.title = title
        self.author = author
        self.book_id = book_id
        self.is_available = True  
    def __str__(self):
        availability = "Available" if self.is_available else "Borrowed"
        return f"Title: {self.title}\nAuthor: {self.author}\nBook ID: {self.book_id}\nStatus: {availability}"
class LibraryCatalog:
    def __init__(self):
        self.books = {}  
    def add_book(self, title, author, book_id):
        if book_id not in self.books:
            new_book = Book(title, author, book_id)
            self.books[book_id] = new_book
            return f"Added '{title}' by {author} (ID: {book_id}) to the library catalog."
        else:
            return f"Book with ID {book_id} already exists in the catalog."

    def borrow_book(self, book_id):
        # Borrow a book if it's available
        if book_id in self.books:
            book = self.books[book_id]
            if book.is_available:
                book.is_available = False
                return f"Borrowed '{book.title}' by {book.author}."
            else:
                return f"The book '{book.title}' by {book.author} is currently unavailable."
        else:
            return f"Book with ID {book_id} is not found in the catalog."

    def return_book(self, book_id):
        # Return a borrowed book to the library
        if book_id in self.books:
            book = self.books[book_id]
            if not book.is_available:
                book.is_available = True
                return f"Returned '{book.title}' by {book.author}."
            else:
                return f"The book '{book.title}' by {book.author} is already available."
        else:
            return f"Book with ID {book_id} is not found in the catalog."

    def display_available_books(self):
        # Display a list of available books
        available_books = [book for book in self.books.values() if book.is_available]
        if available_books:
            available_books_list = "\n\n".join(str(book) for book in available_books)
            return f"Available Books:\n{available_books_list}"
        else:
            return "No books are currently available in the library."
    library = LibraryCatalog()

    # Adding books to the library catalog
    print(library.add_book("The Great Gatsby", "F. Scott Fitzgerald", "B001"))
    print(library.add_book("To Kill a Mockingbird", "Harper Lee", "B002"))
    print(library.add_book("1984", "George Orwell", "B003"))

    # Displaying available books
    print(library.display_available_books())

    # Borrowing a book
    print(library.borrow_book("B002"))

    # Displaying available books after borrowing
    print(library.display_available_books())

    # Returning a book
    print(library.return_book("B002"))

    # Displaying available books after returning
    print(library.display_available_books())
```

```python
#Problem 5: 
#Product Inventory Create classes representing a product and an inventory system. 
#Implement methods to add products to the inventory, update product quantity, and display available products.

class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity
    def __str__(self):
        return f"Product ID: {self.product_id}\nName: {self.name}\nPrice: ${self.price:.2f}\nQuantity: {self.quantity}"
class InventorySystem:
    def __init__(self):
        self.products = {}  
    def add_product(self, product_id, name, price, quantity):
        if product_id not in self.products:
            new_product = Product(product_id, name, price, quantity)
            self.products[product_id] = new_product
            return f"Added '{name}' (ID: {product_id}) to the inventory."
        else:
            return f"Product with ID {product_id} already exists in the inventory."
    def update_product_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            product = self.products[product_id]
            product.quantity = new_quantity
            return f"Updated quantity for '{product.name}' (ID: {product_id}) to {new_quantity}."
        else:
            return f"Product with ID {product_id} is not found in the inventory."
    def display_available_products(self):
        available_products = [product for product in self.products.values() if product.quantity > 0]
        if available_products:
            available_products_list = "\n\n".join(str(product) for product in available_products)
            return f"Available Products:\n{available_products_list}"
        else:
            return "No products are currently available in the inventory."
    inventory_system = InventorySystem()

    # Adding products to the inventory
    print(inventory_system.add_product("P001", "Laptop", 999.99, 5))
    print(inventory_system.add_product("P002", "Smartphone", 499.99, 10))
    print(inventory_system.add_product("P003", "Headphones", 59.99, 20))

    # Displaying available products
    print(inventory_system.display_available_products())

    # Updating product quantity
    print(inventory_system.update_product_quantity("P002", 8))

    # Displaying available products after updating quantity
    print(inventory_system.display_available_products())
```

```python
#Problem 7: 
#Student Management Create a class representing a student with attributes like student ID, name, and grades.
#Implement methods to calculate the average grade and display student details.


class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades
    def calculate_average_grade(self):
        if len(self.grades) > 0:
            return sum(self.grades) / len(self.grades)
        else:
            return 0 
    def display_student_details(self):
        return f"Student ID: {self.student_id}\nName: {self.name}\nGrades: {', '.join(map(str, self.grades))}"
    def __str__(self):
        return self.display_student_details()
    student1 = Student("2010144", "Shaik Afroz", [99, 92, 98, 95, 100])
    average_grade = student1.calculate_average_grade()
    print(student1.display_student_details())
    print(f"Average Grade: {average_grade:.2f}")
```

```python
#Problem 8: 
#Email Management Create a class representing an email with attributes like sender, recipient, and subject. 
#Implement methods to send an email and display email details.

class Email:
    def __init__(self, sender, recipient, subject, message):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.message = message
        self.is_sent = False  
    def send_email(self):
        if not self.is_sent:
            self.is_sent = True
            return f"Email sent from {self.sender} to {self.recipient} with subject '{self.subject}'."
        else:
            return "This email has already been sent."
    def display_email_details(self):
        return f"Sender: {self.sender}\nRecipient: {self.recipient}\nSubject: {self.subject}\nMessage: {self.message}"
    def __str__(self):
        return self.display_email_details()
    email1 = Email("sender@example.com", "recipient@example.com", "Meeting Invitation", "Hello,\nPlease join our meeting tomorrow at 2 PM.")
    # Sending the email
    send_result = email1.send_email()

    # Displaying email details
    print(email1.display_email_details())
    print(send_result)

    # Trying to send the email again
    send_result = email1.send_email()
    print(send_result)
```

```python
#Problem 9: 
#Social Media Profile Create a class representing a social media profile with attributes like username and posts.
#Implement methods to add posts, display posts, and search for posts by keyword.

class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []
    def add_post(self, post_content):
        self.posts.append(post_content)
        return f"Posted: '{post_content}'"
    def display_posts(self):
        if self.posts:
            posts_list = "\n\n".join([f"{i+1}. {post}" for i, post in enumerate(self.posts)])
            return f"Posts by {self.username}:\n{posts_list}"
        else:
            return f"{self.username} has no posts."
    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword in post]
        if matching_posts:
            posts_list = "\n\n".join([f"{i+1}. {post}" for i, post in enumerate(matching_posts)])
            return f"Posts by {self.username} containing '{keyword}':\n{posts_list}"
        else:
            return f"No posts by {self.username} contain '{keyword}'."
    profile1 = SocialMediaProfile("user123")

    # Adding posts to the profile
    print(profile1.add_post("Hello, world!"))
    print(profile1.add_post("Just posted a new picture."))
    print(profile1.add_post("Feeling excited about the weekend!"))

    # Displaying all posts
    print(profile1.display_posts())

    # Searching for posts containing a keyword
    keyword = "excited"
    print(profile1.search_posts(keyword))
```

```python
#Problem 10: 
#ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. 
#Implement methods to add tasks, mark tasks as completed, and display pending tasks.

class ToDoList:
    def __init__(self):
        self.tasks = {}  
    def add_task(self, task, due_date):
        if task not in self.tasks:
            self.tasks[task] = due_date
            return f"Task '{task}' due on {due_date} added to the ToDo list."
        else:
            return f"Task '{task}' already exists in the ToDo list."
    def mark_task_completed(self, task):
        if task in self.tasks:
            del self.tasks[task]
            return f"Task '{task}' marked as completed."
        else:
            return f"Task '{task}' is not found in the ToDo list."
    def display_pending_tasks(self):
        if self.tasks:
            pending_tasks_list = "\n".join([f"Task: {task}, Due: {due_date}" for task, due_date in self.tasks.items()])
            return f"Pending Tasks:\n{pending_tasks_list}"
        else:
            return "No pending tasks in the ToDo list."
    to_do_list = ToDoList()

    # Adding tasks to the ToDo list
    print(to_do_list.add_task("Buy groceries", "2023-10-15"))
    print(to_do_list.add_task("Submit report", "2023-10-20"))
    print(to_do_list.add_task("Call client", "2023-10-18"))

    # Displaying pending tasks
    print(to_do_list.display_pending_tasks())

    # Marking a task as completed
    print(to_do_list.mark_task_completed("Submit report"))

    # Displaying pending tasks after marking a task as completed
    print(to_do_list.display_pending_tasks())
```
